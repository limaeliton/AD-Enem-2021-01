### Join e concat

***

No Pandas `pd.merge()` é uma função utilizada para unir datasets por meio de uma coluna que seja comum para ambos. Frequentemente precisamos consultar dados em várias tabelas e consolidar em uma tabela final. Vamos imaginar que precisamos construir uma tabela com dados cadastrais dos clientes de uma empresa. Provável que estes dados estejam espalhados ao longo de várias tabelas.

Exemplo:

* tb_cadastro
* tb_celular
* tb_endereco
* tb_tempo_relacionamento

Comum a todas as tabelas será provavelmente alguma coluna como o id do cliente (primary key). Desta forma, para consolidar todas as informações cadastrais dos clientes em uma única tabela precisaríamos percorrer várias tabelas e fazer um join. Ocorre ainda a possibilidade de termos que consolidar outros dados por meio de transformação de variáveis e consolidar nesta tabela cadastral, pense por exemplo se a necessidade fosse conta quantas compras este consumidor fez no último mês.

````python
cliente_ocupacao = pd.DataFrame({
        'id_cliente': [1, 2, 3, 4, 5],
        'occupation': ['Data Scientist', 'Economist', 'Data Scientist', 'Data Engineer', 'Data Scientist']
})
````

````python
cliente_experiencia = pd.DataFrame({
          'id_cliente': [2, 3, 4, 5],
          'experience': [10, 3, 2, 4]
})
````

````python
display(cliente_ocupacao,cliente_experiencia)
````

````python
pd.merge(cliente_ocupacao, cliente_experiencia)
````

````python
cliente_merge = pd.merge(cliente_ocupacao, cliente_experiencia , on= 'id_cliente', how='inner')
````

````python
cliente_merge
````

````python
cliente_estado = pd.DataFrame({
        'id_cliente': [2, 3, 5, 8],
        'State': ['BA', 'SP', 'SP', 'RJ']
}) 
````

````python
pd.merge(cliente_merge, cliente_estado, on='id_cliente')
````

````python
cliente_data = pd.DataFrame({
    'id_cliente': [2, 2, 2, 5, 5],
    'order_date': [2019, 2020, 2022, 2017, 2022]
})
````

````python
pd.merge(cliente_merge , cliente_data)
````

````python
cliente_idade =  pd.DataFrame({
    'numero_cliente': [1, 2, 3, 4, 5, 10, 11, 12, 13],
    'age': [40, 33, 27, 35, 39, 44, 25, 23, 29]
})
````

````python
pd.merge(cliente_ocupacao, cliente_idade, left_on='id_cliente', right_on='numero_cliente', how='inner')\
    .drop(columns= 'numero_cliente')
````

````python
cliente_ocupacao_modificado = cliente_ocupacao.set_index('id_cliente')
````

````python
cliente_experiencia_modificado = cliente_experiencia.set_index('id_cliente')
````

````python
pd.merge(
    cliente_ocupacao_modificado,
    cliente_experiencia_modificado,
    left_index= True,
    right_index=True
)\
.reset_index()
````

````python
cliente_ocupacao_modificado.join(cliente_experiencia_modificado, how='inner')
````
